```python
# Define the categories
categories = ["Stewardship", "Interdependence", "Curious Realism", "Sacredness of Nature"]

# Initialize an empty list to store row data
df_data = []

# Generate 1000 rows of data
for _ in range(1000):
    # Generate random integer values for each category
    row_data = {category: random.randint(1, 10) for category in categories}
    # Append the row data to the list
    df_data.append(row_data)

df=pd.DataFrame(df_data)
df.head(10)
```

```python
category_ranges = {
    "Stewardship": (7, 10),
    "Interdependence": (4, 6),
    "Curious Realism": (6, 8),  
    "Sacredness of Nature": (8, 10)  
}

# Initialize an empty list to store row data
df3_data = []

# Generate 1000 rows of data
for _ in range(1000):
    # Generate random integer values for each category within custom ranges
    row_data = {category: random.randint(*category_ranges[category]) for category in categories}
    # Append the row data to the list
    df3_data.append(row_data)

# Now you can create a DataFrame from the list of dictionaries
df3 = pd.DataFrame(df3_data)

df3
```
